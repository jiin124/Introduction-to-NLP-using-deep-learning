# 2. 단층 퍼셉트론(Single-Layer Perceptron)

퍼셉트론은 단층 퍼셉트론과 다층 퍼셉트론으로 나누어지는데, 단층 퍼셉트론은 값을 보내는 단계와 값을 받아서 출력하는 두 단계로만 이루어진다. 이때 이 각 단계를 보통 층(layer)라고 부르면, 이 두개의 층을 입력층(input layer), 출력층(output layer)라고 한다. 

단층 퍼셉트론을 이용하면 and,nand,or게이트를 쉽게 구현할 수 있다. 게이트 연산에 쓰이는 것은 두개의 입력값과 하나의 출력값이다. 예를 들어 and 게이트의 경우 두 개의 입력값이 모두 1인 경우에만 출력값이 1이 나오는 구조를 가지고 있다. 

단층 퍼셉트론의 식을 이용해 and 게이트를 만족하는 두 개의 가중치와 편향 값에는 뭐가 있을까? 각각 w1,w2,b라고 한다면 [0.5,0.5,0.7], [0.5,0.5,-0.8],[1.0,1.0,-1.0] 등 이 외에도 다양한 가중치와 편향의 조합이 나올 수 있다. 이해를 돕기 위해 and 게이트를 위한 매개 변수 값을 가진 단층 퍼셉트론의 식을 파이썬 코드로 간단히 구현할 수 있다.

```python
def AND_gate(x1,x2):
  w1=0.5
  w2=0.5
  b=-0.7
  result=x1*w1+x2*w2+b
  if result<=0:
    return 0
  else:
    return 1
```

```python
AND_gate(0,0),AND_gate(1, 1)
```

```python
def NAND_gate(x1,x2):
  w1=-0.5
  w2=-0.5
  b=0.7
  result=x1*w1+x2*w2+b
  if result<=0:
    return 0
  else:
    return 1
```

```python
NAND_gate(0, 0), NAND_gate(0, 1), NAND_gate(1, 0), NAND_gate(1, 1)
```

```python
def OR_gate(x1,x2):
  w1=0.6
  w2=0.6
  b=-0.5
  result=x1*w1+x2*w2+b
  if result<=0:
    return 0
  else:
    return 1
```

```python
OR_gate(0, 0), OR_gate(0, 1), OR_gate(1, 0), OR_gate(1, 1)
```

위의 함수에 and 게이트 입력값을 모두 넣어보면 오직 두개의 입력값이 1인 경우에만 1을 출력한다. 

단층 퍼셉트론은 and게이트, nand 게이트, or 게이트 또한 구현가능하다. 하지만 단층 퍼셉트론으로 구현이 불가능한 게이트가 있는데 바로 xor게이트이다. XOR게이트는 입력값 두개가 서로 다른 값을 갖고 있을 때에만 출력밧이 1이되고, 입력값 두개가 서로 같은 값을 가지면 출력값이 0이 되는 게이트이다. 위의 파이썬 코드에 아무리 수많은 가중치와 편향을 넣어봐도 XOR 게이트를 구현하는 것은 불가능하다. 

단층 퍼셉트론은 직선 하나로 두 영역을 나눌 수 있는 문제에 대해서만 구현이 가능하기 때문이다.

그림에서 출력값 0을 하얀색 원, 1을 검은색 원으로 표현했다. and 게이트를 충족하려면 하얀색 언과 검은색 원을 직선으로 나누게 된다. 마찬가지로 nand 게이트나 or 게이트에 대해서도 시각화를 했을 때 직선으로 나누는 것이 가능하다.

# 3. 다층 퍼셉트론(MultiLayer Perceptron, MLP)

XOR 게이트는 기존의 and,nand,or 게이트를 조합하면 만들 수 있다. 퍼셉트론 관점에서 말하면, 층을 더 쌓으면 만들 수 있다. 다층 퍼셉트론과 단층 퍼셉트론의 차이는 단층 퍼셉트론은 입력층과 출력층만 존재하지만, 다층 퍼셉트론은 중간에 층을 더 추가하였다는 점이다. 이렇게 입력층과 출력층 사이에 존재하는 층을 은닉층(hidden layer)이라고 한다.


image.png 

즉 다층 퍼셉트론은 중간에 은닉층이 존재한다는 점이 단층 퍼셉트론과 다르다. 다층 퍼셉트론을 줄여서 MLP라고도 부른다. 

XOR 게이트 구현

```python
def XOR_gate(x1,x2):
  n1=NAND_gate(x1,x2)
  n2=OR_gate(x1,x2)
  return AND_gate(n1,n2)
```

```python
XOR_gate(0,0),XOR_gate(0,1),XOR_gate(1,0),XOR_gate(1,1)
```

XOR 예제에서는 은닉층 1개만으로 무제를 해결할 수 있었지만, 다층 퍼셉트론은 중간에 수많은 은닉층을 추가할 수 있다. 은닉층의 개수는 2개일 수도 있고, 수십개일수도있고, 사용자가 설정하기 나름이다. 
image.png

위와 같이 은닉층이 2개 이상인 신경망을 심층 신경망(Deep neural Network,DNN)이라고 한다. 심층 신경망은 다층 퍼셉트론만 이야기하는 것이 아니라, 여러 변형된 다양한 신경망들도 은닉층이 2개 이상이 되면 심층 신경망이라고 한다. 

지금까지는 OR, AND, XOR 게이트 등. 퍼셉트론이 가야할 정답을 참고로 퍼셉트론이 정답을 출력할 때까지 가중치를 바꿔보면서 맞는 가중치를 찾았다.
 즉, 가중치를 수동으로 찾았습니다. 하지만 이제는 기계가 가중치를 스스로 찾아내도록 자동화시켜야하는데, 이것이 머신 러닝에서 말하는 학습(training) 단계에 해당된다. 앞서 선형 회귀와 로지스틱 회귀에서 보았듯이 손실 함수(Loss function)와 옵티마이저(Optimizer)를 사용한다. 그리고 만약 학습을 시키는 인공 신경망이 심층 신경망일 경우에는 이를 심층 신경망을 학습시킨다고 하여, 딥 러닝(Deep Learning)이라고 한다.
